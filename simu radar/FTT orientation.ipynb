```python
import numpy as np
import matplotlib.pyplot as plt
x = np.zeros((128,128))
x[40:41,40:42]=1
l =np.abs( np.fft.fft2(x))
```

```python
plt.figure()
plt.contourf(l)
plt.colorbar()
plt.show()
```
